# Project: Analysis of Seattle Bike Collison - part 1

In 2018, Seattle ranked the 1st bike-friendly city in the US. Free-floating bike share companies have been sprawling out since 2015 and now their colorful bikes are available on any street in the city. Burk-Gillman Trail, one of the longest bike lane connecting Ballard to UW, is always busy by commuters and weekend riders. 

Such popularity of bike riding may also increase the chance of bike incidents which is against the Seattleite's welfare and safety. Regarding the City of Seattle has put its effort to create a bike-friendly environment, understanding the detail of bike collisions occuring in the city should contribute to the endeavor. Motivated by this goal, I came up with questions to answer: How much are bike collisions severe and where do they happen the most? 

To answer the questions, I used the collision data provided by the City of Seattle. In this project, I analyzed the proportion of injuries to all bike incidents, specific addresses and locations that reveal more incident rates than other areas, and finally influences of other factors(i.e., seasonal differences).


## Organizing and cleaning data

I downloaded the dataset from the official website of the City of Seattle (https://data.seattle.gov/Transportation/Collisions/vac5-r8kk). The link to attribute information is here (https://www.seattle.gov/Documents/Departments/SDOT/GIS/Collisions_OD.pdf).

## How much collisions are severe and where they happen the most

In the dataset, severity of collision is categorized by property damage, injury, serious injury, and fatality. 
In general, collisions (i.e., pedestrian, bicycle, vehicle) caused property damage the most: It took approximately 62% of the total. Any level of injury took approximately 28% of the total. 

Then I visualized the locations of collision on the city map. According to the result, the downtown area and major free/highways (i.e., I-5, Route 99, MLK ave, ...) were places that collision caused severe injury the most.

## How much are bike collisions severe?

Bike collisions are not okay. Any level of injury (i.e., injury, serious injury, fatality) took 87% of the total number of bike collision. 

That said, the bike riders have more than 80% chance of getting injured when bike collision happened. 

Interestingly, the locations of bike collisions did not include two major free/highway in Seattle area: I-5 and Route 99.

Except the downtown area, high frequency of bike collision was shown in (1) streets near campus such as U district, Broadway and 12th in Capitol Hill, (2) "Amazon" district including SLU, Mercer St, and Stewert St especially connecting SLU and the downtown, and (3) major roads of Fremont (i.e., Stone way) and Wallingford area.
